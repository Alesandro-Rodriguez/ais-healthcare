# Data Cleaning

```python
# Convert date columns to datetime
df['EmpAnnivDate'] = pd.to_datetime(df['EmpAnnivDate'], errors='coerce')
df['EmpRNExperienceDate'] = pd.to_datetime(df['EmpRNExperienceDate'], errors='coerce')

# Verify the conversion
print(df[['EmpAnnivDate', 'EmpRNExperienceDate']].head())
```

```python
# Fill missing values
df['EmpPositionIsSuper'].fillna('N', inplace=True)
df['MgrName'].fillna('Unknown', inplace=True)
df['EmpFirstName'].fillna('Unknown', inplace=True)

# Check if missing values remain
print(df.isnull().sum())
```

```python
# Calculate tenure in years
df['Tenure'] = (pd.Timestamp.now() - df['EmpAnnivDate']).dt.days / 365

# Check the new column
print(df[['EmpAnnivDate', 'Tenure']].head())
```

```python
# Replace future dates with NaT
df.loc[df['EmpAnnivDate'] > pd.Timestamp.now(), 'EmpAnnivDate'] = pd.NaT

# Recalculate tenure after correction
df['Tenure'] = (pd.Timestamp.now() - df['EmpAnnivDate']).dt.days / 365
```

```python
# Summary of the cleaned dataset
print(df.info())
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())
```

```python
# Count employees by state
state_distribution = df['facility_state'].value_counts()
print(state_distribution)
```

```python
import matplotlib.pyplot as plt


# Plot top 10 states
plt.figure(figsize=(10, 6))
state_distribution.head(10).plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 10 States by Employee Count')
plt.xlabel('State')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Count employees by facility
facility_distribution = df['EmpProcessLevelDesc'].value_counts()
print(facility_distribution)
```

```python
# Plot top 10 facilities
plt.figure(figsize=(10, 6))
facility_distribution.head(10).plot(kind='bar', color='orange', edgecolor='black')
plt.title('Top 10 Facilities by Employee Count')
plt.xlabel('Facility')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot the fixed tenure distribution
plt.figure(figsize=(10, 6))
plt.hist(df['Tenure'], bins=30, edgecolor='black', alpha=0.7, color='green')
plt.title('Distribution of Employee Tenure')
plt.xlabel('Tenure (Years)')
plt.ylabel('Number of Employees')
plt.tight_layout()
plt.show()
```

```python
# Count supervisory roles
supervisory_distribution = df['EmpPositionIsSuper'].value_counts()

# Plot supervisory roles
plt.figure(figsize=(6, 6))
supervisory_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightskyblue'])
plt.title('Supervisory Roles Distribution')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
