# Data Cleaning

```python
# Convert date columns to datetime
df['EmpAnnivDate'] = pd.to_datetime(df['EmpAnnivDate'], errors='coerce')
df['EmpRNExperienceDate'] = pd.to_datetime(df['EmpRNExperienceDate'], errors='coerce')

# Verify the conversion
print(df[['EmpAnnivDate', 'EmpRNExperienceDate']].head())
```

```python
# Fill missing values
df['EmpPositionIsSuper'].fillna('N', inplace=True)
df['MgrName'].fillna('Unknown', inplace=True)
df['EmpFirstName'].fillna('Unknown', inplace=True)

# Check if missing values remain
print(df.isnull().sum())
```

```python
# Calculate tenure in years
df['Tenure'] = (pd.Timestamp.now() - df['EmpAnnivDate']).dt.days / 365

# Check the new column
print(df[['EmpAnnivDate', 'Tenure']].head())
```

```python
# Replace future dates with NaT
df.loc[df['EmpAnnivDate'] > pd.Timestamp.now(), 'EmpAnnivDate'] = pd.NaT

# Recalculate tenure after correction
df['Tenure'] = (pd.Timestamp.now() - df['EmpAnnivDate']).dt.days / 365
```

```python
# Summary of the cleaned dataset
print(df.info())
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())
```

# EDA Expanded

```python
# Count supervisory roles by state
supervisory_by_state = df[df['EmpPositionIsSuper'] == 'Y'].groupby('facility_state')['EmpPositionIsSuper'].count()

# Plot top 10 states with supervisory roles
plt.figure(figsize=(10, 6))
supervisory_by_state.sort_values(ascending=False).head(10).plot(kind='bar', color='purple', edgecolor='black')
plt.title('Top 10 States by Supervisory Roles')
plt.xlabel('State')
plt.ylabel('Number of Supervisory Roles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculate ratio of supervisors to non-supervisors
supervisory_ratio = df['EmpPositionIsSuper'].value_counts(normalize=True) * 100

# Display the ratio
print(f"Supervisory to Non-Supervisory Ratio:\n{supervisory_ratio}")
```

```python
# Calculate average tenure by state
avg_tenure_by_state = df.groupby('facility_state')['Tenure'].mean()

# Plot top 10 states by average tenure
plt.figure(figsize=(10, 6))
avg_tenure_by_state.sort_values(ascending=False).head(10).plot(kind='bar', color='teal', edgecolor='black')
plt.title('Top 10 States by Average Employee Tenure')
plt.xlabel('State')
plt.ylabel('Average Tenure (Years)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculate average tenure by supervisory status
avg_tenure_by_supervisory = df.groupby('EmpPositionIsSuper')['Tenure'].mean()

# Plot average tenure by supervisory status
plt.figure(figsize=(6, 6))
avg_tenure_by_supervisory.plot(kind='bar', color=['red', 'blue'], edgecolor='black')
plt.title('Average Tenure by Supervisory Status')
plt.xlabel('Supervisory Role')
plt.ylabel('Average Tenure (Years)')
plt.tight_layout()
plt.show()
```

```python
# Count total staff and supervisors by facility
facility_staff = df.groupby('EmpProcessLevelDesc')['Emp34Id'].count()
facility_supervisors = df[df['EmpPositionIsSuper'] == 'Y'].groupby('EmpProcessLevelDesc')['EmpPositionIsSuper'].count()

# Merge and calculate ratio
facility_analysis = pd.DataFrame({'Total Staff': facility_staff, 'Supervisors': facility_supervisors})
facility_analysis['Supervisor Ratio'] = (facility_analysis['Supervisors'] / facility_analysis['Total Staff']) * 100
facility_analysis.fillna(0, inplace=True)

# Display top 10 facilities by supervisor ratio
print(facility_analysis.sort_values('Supervisor Ratio', ascending=False).head(10))
```

```python
# Combine tenure and supervisory analysis
low_tenure_facilities = df.groupby('EmpProcessLevelDesc').agg(
    Average_Tenure=('Tenure', 'mean'),
    Supervisors=('EmpPositionIsSuper', lambda x: (x == 'Y').sum()),
    Total_Staff=('Emp34Id', 'count')
)
low_tenure_facilities['Supervisor_Ratio'] = (low_tenure_facilities['Supervisors'] / low_tenure_facilities['Total_Staff']) * 100

# Identify facilities with tenure < 5 years and supervisor ratio < 5%
at_risk_facilities = low_tenure_facilities[(low_tenure_facilities['Average_Tenure'] < 5) &
                                           (low_tenure_facilities['Supervisor_Ratio'] < 5)]
print("Facilities with Low Tenure and Leadership:")
at_risk_facilities.head(100)
```

```python
# Group data by state and calculate the total staff and supervisors
state_analysis = df.groupby('facility_state').agg(
    Total_Staff=('Emp34Id', 'count'),
    Supervisors=('EmpPositionIsSuper', lambda x: (x == 'Y').sum())
)

# Calculate the supervisor ratio
state_analysis['Supervisor_Ratio'] = (state_analysis['Supervisors'] / state_analysis['Total_Staff']) * 100

# Sort the results by Supervisor Ratio and reset index for better readability
state_analysis = state_analysis.sort_values('Supervisor_Ratio', ascending=False).reset_index()

# Display the top 10 states by Supervisor Ratio
print(state_analysis.head(10))

# Plotting Supervisor Ratio by State
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.bar(state_analysis['facility_state'], state_analysis['Supervisor_Ratio'], color='orange', edgecolor='black')
plt.title('Supervisor Ratio by State')
plt.xlabel('State')
plt.ylabel('Supervisor Ratio (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Displaying the data table for all states
import pandas as pd
pd.set_option('display.max_rows', None)  # Remove this line if you don’t need to display all rows
print(state_analysis)
```
